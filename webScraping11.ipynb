```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Function to scrape RSS feeds and return article data
def scrape_rss_feed(url, category, source_name):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'xml')
    articles = []

    # Loop through each item in the RSS feed
    for item in soup.find_all('item'):
        title = item.title.text if item.title else "No Title"
        link = item.link.text if item.link else "No Link"
        pub_date = item.pubDate.text if item.pubDate else "No Publication Date"
        summary = item.description.text if item.description else "No Summary"
        summary = BeautifulSoup(summary, 'html.parser').get_text(strip=True)  # Clean HTML tags from the summary

        # Append each article as a dictionary
        articles.append({
            'Title': title,
            'Summary': summary,
            'Publication Date': pub_date,
            'Source': source_name,
            'Category': category,
            'URL': link
        })

    return articles

# URLs for different categories for BBC and Times of India
rss_feeds = [
    # BBC
    {"url": "http://feeds.bbci.co.uk/news/business/rss.xml", "category": "Business", "source": "BBC"},
    {"url": "http://feeds.bbci.co.uk/news/technology/rss.xml", "category": "Technology", "source": "BBC"},
    {"url": "http://feeds.bbci.co.uk/sport/rss.xml", "category": "Sports", "source": "BBC"},
    {"url": "http://feeds.bbci.co.uk/news/entertainment_and_arts/rss.xml", "category": "Entertainment", "source": "BBC"},
    # Times of India
    {"url": "https://timesofindia.indiatimes.com/rssfeeds/1898055.cms", "category": "Business", "source": "Times of India"},
    {"url": "https://timesofindia.indiatimes.com/rssfeeds/5880659.cms", "category": "Technology", "source": "Times of India"},
    {"url": "https://timesofindia.indiatimes.com/rssfeeds/4719148.cms", "category": "Sports", "source": "Times of India"},
    {"url": "https://timesofindia.indiatimes.com/rssfeeds/1081479906.cms", "category": "Entertainment", "source": "Times of India"}
]

# List to hold all articles
all_articles = []

# Scrape data from each RSS feed
for feed in rss_feeds:
    articles = scrape_rss_feed(feed['url'], feed['category'], feed['source'])
    all_articles.extend(articles)  # Add articles to the list

# Convert the list of articles to a DataFrame
df = pd.DataFrame(all_articles)

# Save the DataFrame to a CSV file
df.to_csv("news.csv", index=False)

# Display the first few rows of the DataFrame
print(df.head())
```

```python
df['CategoryId'] = df['Category'].factorize()[0]
df[25:39]
```

```python
def remove_tags(text):
  remove = re.compile(r'')
  return re.sub(remove, '', text)
df['Summary'] = df['Summary'].apply(remove_tags)
```

```python
def special_char(text):
  reviews = ''
  for x in text:
    if x.isalnum():
      reviews = reviews + x
    else:
      reviews = reviews + ' '
  return reviews
df['Summary'] = df['Summary'].apply(special_char)
```

```python
def convert_lower(text):
   return text.lower()
df['Summary'] = df['Summary'].apply(convert_lower)
df['Summary'][1]
```

```python
def remove_stopwords(text):
  stop_words = set(stopwords.words('english'))
  words = word_tokenize(text)
  return [x for x in words if x not in stop_words]
df['Summary'] = df['Summary'].apply(remove_stopwords)
df['Summary'][1]
```

```python
def lemmatize_word(text):
  wordnet = WordNetLemmatizer()
  return " ".join([wordnet.lemmatize(word) for word in text])
df['Summary'] = df['Summary'].apply(lemmatize_word)
df['Summary'][1]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
x = np.array(df.iloc[:,0].values)
y = np.array(df.CategoryId.values)
cv = CountVectorizer(max_features = 5000)
x = cv.fit_transform(df.Summary).toarray()
print("X.shape = ",x.shape)
print("y.shape = ",y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0, shuffle = True)
print(len(x_train))
print(len(x_test))
```

```python
perform_list = [ ]
```

```python
def run_model(model_name, est_c, est_pnlty):

  mdl=''

  if model_name == 'Logistic Regression':
    mdl = LogisticRegression()

  elif model_name == 'Random Forest':
    mdl = RandomForestClassifier(n_estimators=100 ,criterion='entropy' , random_state=0)

  elif model_name == 'Multinomial Naive Bayes':
    mdl = MultinomialNB(alpha=1.0,fit_prior=True)

  elif model_name == 'Support Vector Classifer':
    mdl = SVC()

  elif model_name == 'Decision Tree Classifier':
    mdl = DecisionTreeClassifier()

  elif model_name == 'K Nearest Neighbour':
    mdl = KNeighborsClassifier(n_neighbors=10 , metric= 'minkowski' , p = 4)

  elif model_name == 'Gaussian Naive Bayes':
    mdl = GaussianNB()

  oneVsRest = OneVsRestClassifier(mdl)

  oneVsRest.fit(x_train, y_train)

  y_pred = oneVsRest.predict(x_test)

  # Performance metrics

  accuracy = round(accuracy_score(y_test, y_pred) * 100, 2)

  # Get precision, recall, f1 scores

  precision, recall, f1score, support = score(y_test, y_pred, average='micro')

  print(f'Test Accuracy Score of Basic {model_name}: % {accuracy}')

  print(f'Precision : {precision}')

  print(f'Recall : {recall}')

  print(f'F1-score : {f1score}')

  # Add performance parameters to list

  perform_list.append(dict([

  ('Model', model_name),

  ('Test Accuracy', round(accuracy, 2)),

  ('Precision', round(precision, 2)),

  ('Recall', round(recall, 2)),

  ('F1', round(f1score, 2))

  ]))
```

```python
run_model('Logistic Regression', est_c=None, est_pnlty=None)
```

```python
# Function to predict category based on the summary
def predict_category(summary):
    if isinstance(summary, str):  # Ensure summary is a string before transforming
        y_pred1 = cv.transform([summary])  # Transform the summary using the fitted CountVectorizer
        yy = classifier.predict(y_pred1)   # Predict using the trained classifier

        # Map the prediction to the corresponding category
        if yy == [0]:
            return "Business News"
        elif yy == [1]:
            return "Tech News"
        elif yy == [2]:
            return "Sports News"
        elif yy == [3]:
            return "Entertainment News"
        else:
            return "Other"
    else:
        return "Unknown"  # Handle any non-string cases (if needed)

# Apply the function to predict categories
dataframe['Predicted Category'] = dataframe['Summary'].apply(predict_category)

# Save the updated DataFrame with predicted categories to a new CSV file
dataframe.to_csv('news_articles_with_predicted_categories.csv', index=False)

# Display the first few rows to verify the predictions
print(dataframe[['Title', 'Summary','Category' ,'Predicted Category']].tail())
```
